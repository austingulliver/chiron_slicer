```python
import glob
from astropy.io import fits
import numpy as np
from astropy.stats import mad_std
print ("All libraries imported successfully!") 
```

```python
files_list= []
# [0] file creation UT
# [1] file formatted_night str
# [2] original file name
#[3]new data
#[4]new header

    
    
# Need to merge HEADERS 


files_list = [] 


for file in glob.glob("*.fits"):
    fits_image_filename = file[file.find("\\")+1:] 

    with fits.open(fits_image_filename) as hdulist:
        
        print ("Fie: " +str(fits_image_filename) + ' - -  - -')
        
        if (len(hdulist)) == 2 :
            try :               
                h_0 = hdulist[0].header  #['DATE'].strip().lower()
                h_1 = hdulist[1].header
            except:
                raise Exception ("This hdulist did not have 2 hdu. ")

            if len(h_0) < len(h_1) :
                new_header = h_1
                to_remove_header= h_0
                new_data= hdulist[1].data
            else:
                new_header = h_0
                to_remove_header= h_1
                new_data= hdulist[0].data


            #appending to new header
            for arg in to_remove_header.items():
                new_header.append(arg)
                
        elif (len(hdulist)) == 1 :
            new_header=hdulist[0].header
            new_data = hdulist[0].data
        
        else:
            raise Exception (" Not HDU found")
            
            
        
       
        h_date= new_header['DATE'].strip().lower()
        

        time_i= h_date.index('t')

        night= (h_date[: time_i]).split('-')

        night = [ n.strip() for n in night]

        year = (night[0])[2:]
        month =night[1]
        day =night[2]
        dateStr = str(year)+ str(month)+str(day)



        time= h_date[time_i+1:]        
        times =time.split(':')
        times = [  t.strip() for t in times]
        time = int(times[0]+times[1]+times[2])


        files_list.append([time, dateStr,fits_image_filename,new_data, new_header   ])
                         #  int : time , str : date YYMMDD ,  str : original file name

        
        
for e in files_list:
    print (e[0])

print  ("The file list with all files has a length of " +str(len(files_list)))
        
                 

            
            
```
